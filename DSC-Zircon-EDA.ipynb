## DS1 Visuals


```python
plt.figure(figsize=(10, 6))
sns.histplot(df1['AGE(MA)'].dropna(), bins=20, kde=True)
plt.title('Histogram of Age (Ma) in df1')
plt.xlabel('Age (Ma)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Comment', y='Hf', data=df1)
plt.title('Boxplot of Hf in df1 by Comment')
plt.xlabel('Comment')
plt.ylabel('Hf')
plt.show()
```

```python
# Check if there are any rows where the 'Comment' column is 'Barren'
barren_data = df1[df1['Comment'] == 'Barren']

# Print the first few rows of the filtered DataFrame
print(barren_data.head())
```

```python
import numpy as np

# Replace NaN values with median of 'Cu_tonnage_Mt' column
median_cu_tonnage = df1['Cu_tonnage_Mt'].median()
df1['Cu_tonnage_Mt'] = df1['Cu_tonnage_Mt'].fillna(median_cu_tonnage)

# Plot the boxplot after handling missing values
plt.figure(figsize=(10, 6))
sns.boxplot(x='Comment', y='Cu_tonnage_Mt', data=df1)
plt.title('Boxplot of Cu_tonnage_Mt by Comment in df1')
plt.xlabel('Comment')
plt.ylabel('Cu_tonnage_Mt')
plt.show()
```

```python
selected_vars = ['AGE(MA)', 'Ti', 'La', 'Ce', 'Nd', 'Sm', 'Eu', 'Gd']
sns.pairplot(df1[selected_vars])
plt.suptitle('Scatterplot Matrix of Selected Variables in df1', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = df1[selected_vars].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Corre lation Matrix of Selected Variables in df1')
plt.show()
```
